## Date and year

```python
# Checking the type of data we are dealing with

df_us_clean['date_announced'].dtype
```

```python
# Changing the data time to more easily work with it

df_us_clean['date_announced'] = pd.to_datetime(df_us_clean['date_announced'])
```

```python
# Create a new column

df_us_clean['year'] = df_us_clean['date_announced'].dt.year
df_us_clean['year']
```

```python
# Take a portion of the dataset and create a reduce dataset with only the columns of interest

df_us_clean_year = df_us_clean.groupby('year').count().reset_index()
df_us_clean_year_reduced = df_us_clean_year[['year', 'sdc_deal_no']].rename(columns = {'sdc_deal_no':'count'})
df_us_clean_year_reduced
```

```python
# Visualizing the distribution of deals made per year

plt.figure(figsize=(15,5))
plt.yticks(fontsize=10)
sns.set_style("ticks")
ax = sns.barplot(data=df_us_clean_year_reduced, x='year', y='count', color='#17B3C8')
ax.set(xlabel='Year', ylabel='Number of Deals made')
sns.despine(bottom=True, left=True)
ax.grid(False)
ax.tick_params(bottom=False, left=True)
for _,s in ax.spines.items():
    s.set_color('black')
```

## Investigating Acquirors

```python
# Creating a new reduced dataset to investigate the number of acquisitions and mergers performed by an Acquiror

df_us_clean_acquiror = df_us_clean.groupby('acquiror_full_name',sort=False).count().reset_index().sort_values(by='sdc_deal_no',ascending=False)
df_us_clean_acquiror = df_us_clean_acquiror[['acquiror_full_name', 'sdc_deal_no']].rename(columns = {'sdc_deal_no':'count'})
df_us_clean_acquiror_filtered = df_us_clean_acquiror[df_us_clean_acquiror['count'] > 50]
df_us_clean_acquiror_filtered
```

```python
# Visualizing the distribution of deals made by a company

plt.figure(figsize=(15,7))
plt.xticks(fontsize=10, rotation=90)
plt.yticks(fontsize=10)
sns.set_style("ticks")
ax = sns.barplot(data=df_us_clean_acquiror_filtered, x='acquiror_full_name', y='count', color='#17B3C8')
ax.set(xlabel='Company', ylabel='Number of Deals made')
sns.despine(bottom=True, left=True)
ax.grid(False)
ax.tick_params(bottom=False, left=True)
for _,s in ax.spines.items():
    s.set_color('black')
```

```python
# Creating a new reduced dataset to investigate the number of acquisitions and mergers performed in which industry

df_us_clean_acquiror_industry = df_us_clean.groupby('acquiror_TRBC_industry',sort=False).count().reset_index().sort_values(by='sdc_deal_no',ascending=False)
df_us_clean_acquiror_industry = df_us_clean_acquiror_industry[['acquiror_TRBC_industry', 'sdc_deal_no']].rename(columns = {'sdc_deal_no':'count', 'acquiror_TRBC_industry': 'acquiror_industry'})
df_us_clean_acquiror_industry_reduced = df_us_clean_acquiror_industry[df_us_clean_acquiror_industry['count'] > 200]
df_us_clean_acquiror_industry_reduced
```

```python
# Visualizing the distribution of deals made by industry

plt.figure(figsize=(15,7))
plt.xticks(fontsize=10, rotation=90)
plt.yticks(fontsize=10)
sns.set_style("ticks")
ax = sns.barplot(data=df_us_clean_acquiror_industry_reduced, x='acquiror_industry', y='count', color='#17B3C8')
ax.set(xlabel='Company', ylabel='Number of Deals made')
sns.despine(bottom=True, left=True)
ax.grid(False)
ax.tick_params(bottom=False, left=True)
for _,s in ax.spines.items():
    s.set_color('black')
```

```python
# Extracting a reduced dataset with the target nation being outside of the US

df_us_clean_target_nation = df_us_clean.groupby('target_nation',sort=False).count().reset_index().sort_values(by='sdc_deal_no',ascending=False)
df_us_clean_target_nation = df_us_clean_target_nation[['target_nation', 'sdc_deal_no']].rename(columns = {'sdc_deal_no':'count'})
df_us_clean_target_nation_foreign = df_us_clean_target_nation[df_us_clean_target_nation['target_nation'] != 'United States']
df_us_clean_target_nation_foreign_reduced = df_us_clean_target_nation_foreign[df_us_clean_target_nation_foreign['count'] > 50]
df_us_clean_target_nation_foreign_reduced
```

```python
# Visualizing the distribution of deals made by Nations outside of the US

plt.figure(figsize=(15,7))
plt.xticks(fontsize=10, rotation=90)
plt.yticks(fontsize=10)
sns.set_style("ticks")
ax = sns.barplot(data=df_us_clean_target_nation_foreign_reduced, x='target_nation', y='count', color='#17B3C8')
ax.set(xlabel='Nation', ylabel='Number of Deals made')
sns.despine(bottom=True, left=True)
ax.grid(False)
ax.tick_params(bottom=False, left=True)
for _,s in ax.spines.items():
    s.set_color('black')
```
