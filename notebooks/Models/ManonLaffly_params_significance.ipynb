```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv('../MA_PREDICTOR/data/ma_data_car_clean.csv')
```

```python
data.head()
```

```python
y=data[['car']]
```

```python
X=data[['consideration_offered', 'shares_acquired', 'shares_at_announcement', 'month', 'cross_border', 'relatedness',
       'economic_sector_ac', 'business_sector_ac', 'economic_sector_target', 'target_status',
       'business_sector_target']]
```

```python
model = smf.ols(formula='car ~ C(consideration_offered) + C(shares_acquired) + C(shares_at_announcement) + month + C(cross_border) + C(relatedness) + C(economic_sector_ac) + C(business_sector_ac) + C(economic_sector_target) + C(target_status) + C(business_sector_target)', data=data).fit()
```

```python
model.rsquared
```

```python
model.params
```

```python
model.summary()
```

```python
p_values=model.pvalues
```

```python
p_values[p_values < 0.05]
```

```python
p_values[p_values < 0.1]
```
