```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../MA_PREDICTOR/data/ma_data_car_clean.csv')
```

```python
data.drop('month', inplace=True, axis=1)
```

```python
horizons = [1, 3, 5, 10]
for horizon in horizons:
    data = data[(abs(data[f'car_{horizon}']) <= 0.3)]
```

We investigate the car_1 distribution of the business sector which represent less than 100 deals in our dataframe. To do that, we group them under the business sector's name "others" and we plot a catplot graph. 

```python
small_activity = list(data['business_sector_ac'].value_counts()[data['business_sector_ac'].value_counts() < 100].index)
```

```python
small_activity
```

```python
data['business_sector_ac'].replace({'Collective Investments': "others",
                                     'Renewable Energy':"others",
                                     'Academic & Educational Services':"others",
                                     'Investment Holding Companies':"others",
                                     'Uranium':"others"}, inplace=True)
```

```python
plt.figure(figsize=(45, 15))
sns.catplot(data=data, x='business_sector_ac', y='car_1', kind='box', showfliers=False,
           height=5,
           aspect=4);
plt.xticks(rotation=45);
```

The last box represent this "others" sector and it seems that it does not have a different car distribution compared to the sectors that are more represented. Therefore, we can't assume that the deals that occur in minor sector have more probabilities to be successful.
