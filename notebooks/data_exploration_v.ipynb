## Date and year

```python
df_us_filtered_clean['date_announced'].dtype
```

```python
df_us_filtered_clean['date_announced'] = pd.to_datetime(df_us_filtered_clean['date_announced'])
```

```python
df_us_filtered_clean['year'] = df_us_filtered_clean['date_announced'].dt.year
df_us_filtered_clean['year']
```

```python
df_us_filtered_clean_year = df_us_filtered_clean.groupby('year').count()
df_us_filtered_clean_year = df_us_filtered_clean_year.reset_index()
df_us_filtered_clean_year = df_us_filtered_clean_year[['year', 'sdc_deal_no']].rename(columns = {'sdc_deal_no':'count'})
df_us_filtered_clean_year
```

```python
df_us_filtered_clean[['year','sdc_deal_no']].groupby('year').count().plot(kind='bar')
```

```python
#sns.barplot(data=df_us_filtered_clean_year,x = 'Year', y='SDC_Deal_No')
```

```python
df_us_filtered_clean_acquiror = df_us_filtered_clean.groupby('acquiror_full_name',sort=False).count()
df_us_filtered_clean_acquiror = df_us_filtered_clean_acquiror.reset_index().sort_values(by='sdc_deal_no',ascending=False)
df_us_filtered_clean_acquiror = df_us_filtered_clean_acquiror[['acquiror_full_name', 'sdc_deal_no']]
df_us_filtered_clean_acquiror.head(11)
```

```python
df_us_filtered_clean[['acquiror_full_name','sdc_deal_no']].groupby('acquiror_full_name').count().sort_values(by='sdc_deal_no',ascending=False).head(11).plot(kind='bar')
```

```python
df_us_filtered_clean_target_nation = df_us_filtered_clean.groupby('target_nation',sort=False).count()
df_us_filtered_clean_target_nation = df_us_filtered_clean_target_nation.reset_index().sort_values(by='sdc_deal_no',ascending=False)
df_us_filtered_clean_target_nation = df_us_filtered_clean_target_nation[['target_nation', 'sdc_deal_no']]
df_us_filtered_clean_target_nation.head(11)
```

```python
plt.figure(figsize=(40, 16))
sns.barplot(data=df_us_filtered_clean_target_nation.head(11),x='target_nation', y='sdc_deal_no')
```

```python
df_us_filtered_clean_acquiror_industry = df_us_filtered_clean.groupby('acquiror_TRBC_industry',sort=False).count()
df_us_filtered_clean_acquiror_industry = df_us_filtered_clean_acquiror_industry.reset_index().sort_values(by='sdc_deal_no',ascending=False)
df_us_filtered_clean_acquiror_industry = df_us_filtered_clean_acquiror_industry[['acquiror_TRBC_industry', 'sdc_deal_no']]
df_us_filtered_clean_acquiror_industry.head(11)
```
